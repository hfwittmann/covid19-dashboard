```python
import pandas as pd
from subs.getData import getTimeseries_T
```

```python
import yaml
with open('config/run.yml', 'r') as f:
    config = yaml.safe_load(f)
```

```python
timeseries_T = pd.read_json(getTimeseries_T('Death')['timeseries_T'])
timeseries_T['date'] = timeseries_T['date'].apply(lambda x: x.strftime(config['DateFormat']))
timeseries_T
```

```python
countries_selection = ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola',
       'Antigua and Barbuda', 'Argentina', 'Armenia', 'Australia',
       'Austria', 'Azerbaijan', 'Bahrain', 'Bangladesh', 'Barbados',
       'Belarus', 'Belgium', 'Belize', 'Benin', 'Bhutan', 'Bolivia',
       'Bosnia and Herzegovina', 'Botswana']
```

```python
timeseries_T_country_selection = timeseries_T[timeseries_T.country.isin(countries_selection)].copy()
```

```python
timeseries_T_country_selection
```

```python
import plotly
import plotly.express as px
```

```python
px.scatter(
            timeseries_T_country_selection,
            x='T',
            y='cases',
            color='country',
            hover_data = ['date'],
            log_y=True
            )
```
